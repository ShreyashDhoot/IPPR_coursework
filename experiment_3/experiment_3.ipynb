1. Apply average filtring to image for 4 varying mask size 
2. Insert gaussian noise in image and apply mean filtering and median filtering 
3. Insert salt pepper noise in image and apply mean and median filtering 

```python
import cv2
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
img=cv2.imread("D:\ippr\experiment_1\image.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

### Average filtering using different masks 

```python
kernel_size=[3,7,9,16,32]
plt.figure(figsize=(12,9))
for i,kernel in enumerate(kernel_size):
    mask_3=cv2.blur(img,(kernel,kernel))
    plt.subplot(1,(len(kernel_size)),i+1)
    plt.imshow(mask_3,cmap='gray')
    plt.title(f"filter size {kernel} x {kernel}")
    
plt.tight_layout()
plt.show()
```

```python
## Adding Gaussian Noise to the image 
# creates an array of noise of shape image and initialized with 0 
noise=np.zeros(img.shape,np.uint8)
#randn samples numbers from normal distribution randomly 
cv2.randn(noise,mean=50,stddev=50)
# adds this noise to image 
image_gauss=cv2.add(img,noise)
```

```python
# applying median filter 
median_filter=cv2.medianBlur(image_gauss,3)
```

```python
#applying gaussian filter 
gaussian_filter=cv2.GaussianBlur(image_gauss, # <== original image 
                                 (3,3), # <== kernel size
                                 10) # <== std deviation of gaussian filter in X direction 
```

```python
plt.figure(figsize=(12,9))
plt.subplot(1,4,1)
plt.imshow(img,cmap='gray')
plt.title('original image')
plt.subplot(1,4,2)
plt.imshow(image_gauss,cmap='gray')
plt.title('with gaussian noise')
plt.subplot(1,4,3)
plt.imshow(median_filter,cmap='gray')
plt.title('median filtered image')
plt.subplot(1,4,4)
plt.imshow(gaussian_filter,cmap='gray')
plt.title('gaussian filtered image')
```
